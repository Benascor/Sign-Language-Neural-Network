## Introduction

The intended problem is the recognition and live translation of American Sign Language into text, using a Leap Motion IR Camera sensor. A implementation of such a program was developed by a team we were members of during the 2017 RAMhack VR/AR Hackathon, and shows a lot of room for improvement with more advanced methods. We aim to improve this through the use of a neural network in order to improve recognition accuracy.

We are interested in a neural network because they're generally considered to be quite good at pattern recognition once trained, and the letters in american sign language are very much a pattern we're simply trying to match input data to.

The question will ultimately be whether we will be able to provide the network the data pertaining to orientation and placement of hands in a meaningful way. The sensor we will be using to generate data is imperfect, and is highly limited and may itself prove to be the ultimate restriction of the project.

We are interested in the problem and algorithm in part because we would love to have the program we built for the hackathon be more functional and potentially even of use to ASL users. Beyond that, we see a lot of potential use for neural networks and would like to expand our experience with working with one.

## Methods

The core problem at this time is the limited capacity of the system (written in C#) to match to signs. Of the signs implemented there is no significant problems, but further letters in the alphabet proved problematic in some cases as the rudimentary comparisons being made are not accurate or different enough to differentiate some letters. We aim to improve much of this inaccuracy through transitioning from a simple comparison with a difference threshhold to a system based on a trained neural network which results in continuous values.

In order to accomplish this, we currently plan to interface with the Leap Motion camera's python 'Orion' SDK, and will likely implement some portion of an already developed neural net (such as what was written and used as an example by Professor Chuck Anderson in class). The remaining code will be written to transform the data we obtain from the sensor into information we'll be able to input into the neural net.


Our data-set currently will have to be generated by the team/our efforts, as we have yet to find a usable dataset online.

## Possible Results

Assuming we can get some level of consistency on recognition of new data, we will need to examine the data and determine where in the continuous scale we will need to cut off as being too inaccurate/too far off of our target in order to set when a sign is recognized vs a simple hand position.

We hope to be able to get sufficient threshholds and consistency to accurately be able to recognize all the letters of the english alphabet. Whether that proves possible will result on whether the Leap Motion Camera is able to manage to give unique data from some of the more similar hand positions and if the neural net will be able to learn to differentiate between those differences (should they be present).
